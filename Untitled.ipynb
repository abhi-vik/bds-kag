```python
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.metrics import roc_auc_score, precision_score, recall_score, f1_score
from sklearn.feature_selection import RFECV
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV, StratifiedKFold, train_test_split
from sklearn.decomposition import PCA

import lightgbm as lgbm
import xgboost as xgb

from imblearn.over_sampling import SMOTE

from IPython.display import display

sns.set()
sns.set(rc={'figure.figsize':(18,12)})
```

```python
df = pd.read_csv('data/train_final.csv', index_col='Id')
df_ult = pd.read_csv('data/test_final.csv', index_col='Id')

X = df.drop('Y', axis='columns').values
y = df['Y'].values

X_ult = df_ult.values

scale_pos_weight = len(y[y == 0])/len(y[y == 1])
```

```python
model = xgb.XGBRegressor(**{
    'objective': 'binary:logistic',
    'eval_metric': 'auc',
    'max_depth': 11,
    'n_estimators': 10000,
    'learning_rate': 0.075,
    'n_jobs': 8
})

model.fit(X, y)
```

```python
xgb.plot_importance(model);
```

```python
const_params = {
    'objective':                      'binary:logistic', 
#     'booster':                        'gbtree',
    'tree_method':                    'exact', 
    'eval_metric':                    'auc',
#     'scale_pos_weight':               scale_pos_weight
}

tuning_parameters = {
    'max_depth':                      [13], 
    'learning_rate':                  [0.08], 
    'n_estimators':                   [210, 230], 
    'reg_alpha':                      [0], 
    'reg_lambda':                     [1], 
    'colsample_bytree':               [1], 
    'colsample_bylevel':              [0.03], 
    'colsample_bynode':               [0.86], 
#     'random_state': [0]
}

grid = GridSearchCV(xgb.XGBRegressor(**const_params), 
                          tuning_parameters, 
                          cv=StratifiedKFold(5), 
                          scoring='roc_auc', 
                          n_jobs=-1)

grid.fit(df[['f14', 'f15', 'f1', 'f16', 'f8', 'f4', 'f13', 'f23', 'f19']].values, y)

print(f'Best parameters {grid.best_params_}.')
print(f'Best auc score is {grid.best_score_}.')
```
